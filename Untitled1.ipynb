```python
import requests
import json
import pandas as pd
import numpy as np
import re
import math

response = requests.get('https://care.personcentredsoftware.com/mcm/api/v1/284748bf-6409-4367-887d-795f5458e954/organisationapi/Get/serviceusers?includeContacts=true').text;
response_info = json.loads(response)
```

```python
firstNames = []
lastNames = []
userRooms = []
telephones = []
email = []
notes = []
emergencyName=[]

def getUserInformation(): 
    for serviceUser in response_info[0]['serviceUsers']:
        for relationship in serviceUser['relationships']:
            firstName = serviceUser['firstName']
            lastName = serviceUser['lastName']
            userRoom = serviceUser['room']
            relationshipPhones = relationship['telephone']
            relationshipEmail = relationship['email']
            relationshipNotes = relationship['notes']
            relationshipName=relationship['personName']
            
            firstNames.append(firstName)
            lastNames.append(lastName)
            userRooms.append(userRoom)
            telephones.append(relationshipPhones)
            email.append(relationshipEmail) 
            notes.append(relationshipNotes)
            emergencyName.append(relationshipName)
            
```

```python
getUserInformation()
```

```python
flex_df = pd.DataFrame(columns=['Resident Last Name','Resident First Name',
                                'Resident Room','Units','Contacts Last Name','Contacts First Name',
                                'Phones','Notes','Contacts Home Phone','Contacts Mobile Phone','Contacts Email'])
```

```python
flex_df['Resident First Name'] = firstNames
flex_df['Resident Last Name'] = lastNames
flex_df['Resident Room'] = userRooms
flex_df['Phones'] = telephones
flex_df['Contacts First Name'] = emergencyName
flex_df['Contacts Email'] = email
flex_df['Notes'] = notes
```

```python
def telephoneChecker():
   
    phone_data1 = []
    phone_data2 = []
    mobile_list = []
    landline_list = []

    for index, row in flex_df.iterrows():
        if len(row['Phones']) > 1: 
            phone_data1.append(row['Phones'][0])
            phone_data2.append(row['Phones'][1])      
        elif len(row['Phones']) == 1:

            if len(row['Phones'][0]) > 0:
                phone_data1.append(row['Phones'][0])
                phone_data2.append(0) 
            else:
                phone_data1.append(0)
                phone_data2.append(0)

    flex_df["Phone1"] = phone_data1
    flex_df["Phone2"] = phone_data2

    del flex_df['Phones']
    
    for index, row in flex_df.iterrows():
        # 0,0
        if row['Phone1'] == 0:
            mobile_list.append("")
            landline_list.append("")
        # Mobile + ?
        elif re.search("^07", row['Phone1']) is not None:
            if row['Phone2'] == 0:
                # Mobile + 0
                mobile_list.append(row['Phone1'])
                landline_list.append("")
            # Mobile + Landline (or Mobile2)
            else:
                mobile_list.append(row['Phone1'])
                landline_list.append(row['Phone2'])  
        else:
            # Landline + ?
            if row['Phone2'] == 0:
                # Landline + 0
                landline_list.append(row['Phone1'])
                mobile_list.append("")
            # Landline + Mobile (or Landline2)
            else:
                landline_list.append(row['Phone1'])
                mobile_list.append(row['Phone2'])
    
    flex_df['Contacts Mobile Phone'] = mobile_list
    flex_df['Contacts Home Phone'] = landline_list
    del flex_df['Phone1']
    del flex_df['Phone2']
```

```python
def emailChecker():
    
    email_list=[]

    for email in flex_df['Contacts Email']:
        if len(email[0])==0:
            email_list.append("")
        else:
            email_list.append(email[0])
            
    del flex_df['Contacts Email']
    flex_df['Contacts Email']= email_list         
```

```python
telephoneChecker()
emailChecker()
```

```python
def notesChecker():
    
    PR_T = "PR:T"
    PR_V = "PR:V"
    PR_E = "PR:E"
    PR_N = "PR:N"
    
    notes_list=[]
    
    for index, row in flex_df.iterrows():
        # Check Notes that contain None
        if row['Notes'] is None:
            row['Notes'] = ""
#       PR:T
        elif PR_T in row['Notes']:
            row['Notes'] = PR_T
#       PR:V
        elif PR_V in row['Notes']:
            row['Notes'] = PR_V        
#       PR:E
        elif PR_E in row['Notes']:
            row['Notes'] = PR_E
#       PR:N
        elif PR_N in row['Notes']:
            row['Notes'] = "PR_N"            
        else:
            row['Notes'] = ""
notesChecker()
```
